```python
dataset=pd.read_csv("D:\Data_Sets\Brain Stroke.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
```

```python
dataset["gender"]=lb.fit_transform(dataset["gender"])
dataset["ever_married"]=lb.fit_transform(dataset["ever_married"])
dataset["work_type"]=lb.fit_transform(dataset["work_type"])
dataset["Residence_type"]=lb.fit_transform(dataset["Residence_type"])
dataset["smoking_status"]=lb.fit_transform(dataset["smoking_status"])
```

```python
dataset
# here in [gender] 1=Male and 0=female
# here in [ever_married] 1=yes and 0=no
# here in [work_type] 0=Govt_job, 1=private and 2=self_employed
# here in [Residence_type] 1=urban and 0=rural
# here in [smoking_status] 0= , 1=formarly_smoked, 2=never_smoked, 3=smokes
```
